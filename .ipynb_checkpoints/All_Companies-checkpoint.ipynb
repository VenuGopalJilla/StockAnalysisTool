```python
import pandas as pd


df_final = pd.read_csv("C:\\Users\\venu\\Desktop\\Stock Market Analysis\\Data\\Models_Results\\All_Companies_Final_Results.csv")
df_final
```

```python
df_equity = pd.read_csv("C:\\Users\\venu\\Desktop\\Stock Market Analysis\\Data\\Equity.csv")
security_numbers = df_equity["Security Code"].tolist()
security_names = df_equity["Security Name"].tolist()
companies = {str(k) : v for (k, v) in list(zip(security_numbers, security_names))}
companies["542602"] = "Embassy Office Parks REIT"
```

```python
final_columns = ['Company', 'Model', 'Method', 'Percentage']
df1 = pd.DataFrame(columns = final_columns)
df1
```

```python
%%time

for key, value in companies.items():
    name = key + "-" + companies[key]
    df2 = df_final[df_final['Company'] == name]
    df1 = pd.concat([df1, df2.head(1)])

df1
```

```python
df1 = df1.sort_values(by = ['Percentage'], ascending = [False])
df1
```

```python
df1.to_csv('C:\\Users\\venu\\Desktop\\Stock Market Analysis\\Data\\Models_Results\\All_Companies_Final_Results_Single_Value.csv', index = False) 
```
